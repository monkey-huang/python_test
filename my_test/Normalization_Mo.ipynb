```python
from sklearn import preprocessing #標準化數據模組
import numpy as np
import warnings
# 將資料分割成train與test
from sklearn.model_selection import train_test_split
# 生成適合做classification資料
from sklearn.datasets import make_classification 
# Support Vector Machine中的Support Vector Classifier
from sklearn.svm import SVC 
# 可視化數據的模組
import matplotlib.pyplot as plt 

warnings.simplefilter(action='ignore',category=FutureWarning)
```

```python
X, y = make_classification(
    n_samples=300, n_features=2,
    n_redundant=0, n_informative=1, 
    random_state=22, n_clusters_per_class=1, 
    scale=100)
#n_samples樣本的數量,n_features特徵的數量,n_redundant 冗餘特徵數量, 
#n_informative 信息特徵的數量。每個類由多個高斯簇組成，每個高斯簇位於維度n_informative的子空間中的超立方體的頂點周圍。
#random_state ,確定數據集創建的隨機數生成。傳遞一個int，用於跨多個函數調用的可重現輸出。
#n_clusters_per_class 每個類的簇數, scale 按指定值乘以要素。

#y是label資料
#x是實際資料

plt.scatter(X[:, 0], X[:, 1], c=y)#[:,0]每一列的第0行資料
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)#test_size切成30%
clf = SVC()
clf.fit(X_train, y_train)
print(clf.score(X_test, y_test))
```

```python
X = preprocessing.scale(X)
plt.scatter(X[:, 0], X[:, 1], c=y)#[:,0]每一列的第0行資料
plt.show()

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
clf = SVC()
clf.fit(X_train, y_train)
print(clf.score(X_test, y_test))
# 0.9
```
